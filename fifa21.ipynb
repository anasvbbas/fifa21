<h3>separated year, month and day from joined column</h3>

```python
fifa_df['Joined'] = pd.to_datetime(fifa_df['Joined'], format='%b %d, %Y', errors='coerce')
```

```python
fifa_df['day'] = fifa_df['Joined'].dt.day
fifa_df['month'] = fifa_df['Joined'].dt.month
fifa_df['Year'] = fifa_df['Joined'].dt.year
```

```python
fifa_df = fifa_df.drop('Joined', axis=1)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
```

<h3>Tranform the Value, wage and release clause columns into columns of integer</h3>

```python
def konversi_Value(Value):
    Value = Value.replace('€', '') 
    if 'M' in Value:
        return float(Value.replace('M', '')) * 1e6
    elif 'K' in Value:
        return float(Value.replace('K', '')) * 1e3
    else:
        return float(Value)

# Menerapkan fungsi konversi pada kolom 'Harga'
fifa_df['Value'] = fifa_df['Value'].apply(konversi_Value).astype(int)
```
